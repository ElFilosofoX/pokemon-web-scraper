```python
import pandas
import json
import re

def clean_string(s):
    """
    Function to clean special characters from a string.
    """
    s = re.sub(r'[^\x00-\x7F]+', '', s)
    return s.strip()
```

```python
with open('moves_enriched.json', 'r') as file1:
    data = json.load(file1)

cleaned_list = []
for json_obj in data:
    cleaned_obj = {}
    for key, value in json_obj.items():
        cleaned_value = clean_string(value)
        cleaned_obj[key] = cleaned_value
    cleaned_list.append(cleaned_obj)
cleaned_list
```

```python
with open("cleaned_moves_enriched.json", mode='w', encoding='utf-8') as output_file:
        json.dump(cleaned_list, output_file, indent=4)
```
